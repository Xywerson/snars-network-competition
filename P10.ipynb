```python
import os
import re
import time
import glob
import numpy as np
import pandas as pd

import networkx as nx

import matplotlib.pyplot as plt

BASE_DIR = r"C:\Users\Ksawery\Desktop\Classes 3\Social Networks\P10\snars-network-competition"
COMP_DIR = os.path.join(BASE_DIR, "competition")
OUT_DIR  = os.path.join(BASE_DIR, "results_Ksawery")

os.makedirs(OUT_DIR, exist_ok=True)



SEED = 42
np.random.seed(SEED)

print("BASE_DIR:", BASE_DIR)
print("COMP_DIR:", COMP_DIR)
print("OUT_DIR :", OUT_DIR)
```

```python
#helpers like listing files, loading, parsing K/UNC etc....

def list_competition_files(comp_dir=COMP_DIR):
    files = []
    if os.path.isdir(comp_dir):
        files = sorted(glob.glob(os.path.join(comp_dir, "*.csv")))
        if files:
            return files


def load_adjacency_matrix(csv_path):
    A = pd.read_csv(csv_path, header=None).values.astype(float)
    if A.shape[0] != A.shape[1]:
        raise ValueError(f"Adjacency matrix must be square. Got {A.shape} for {csv_path}")

    A[A < 0] = 0.0
    A = 0.5 * (A + A.T)
    return A


def file_category_and_known_k(filepath):
    base = os.path.basename(filepath)

    m = re.search(r"K\s*=\s*(\d+)", base)
    
    if m:
    
        return ("KNOWN_K", int(m.group(1)))
    
    if "UNC" in base.upper():
    
        return ("UNC", None)
    
    return ("UNC", None)


def adjacency_to_graph(A, weight_threshold=0.0):
    
    n = A.shape[0]
    G = nx.Graph()
    G.add_nodes_from(range(n))
    
    rows, cols = np.where(np.triu(A, 1) > weight_threshold)
    
    for i, j in zip(rows, cols):
        w = float(A[i, j])
        if w > weight_threshold:
            G.add_edge(i, j, weight=w)
    
    return G


def basic_graph_stats(A):
    n = A.shape[0]
    edges = np.sum(np.triu(A, 1) > 0)
    density = (2 * edges) / (n * (n - 1)) if n > 1 else 0.0
    deg = np.sum(A > 0, axis=1)
    return {
        "N": int(n),
        "edges": int(edges),
        "density": float(density),
        "min_deg": float(np.min(deg)),
        "max_deg": float(np.max(deg)),
        "avg_deg": float(np.mean(deg)),
    
    }


files = list_competition_files(COMP_DIR)
print("Found files:")
for f in files:
    
    cat, k = file_category_and_known_k(f)
    
    print(" -", os.path.basename(f), "|", cat, "| K:", k)
```

```python
#Visualize all graphs as they are

for f in files:
    A = load_adjacency_matrix(f)
    stats = basic_graph_stats(A)
    G = adjacency_to_graph(A, weight_threshold=0.0)



    plt.figure(figsize=(6, 5))

    pos = nx.spring_layout(G, seed=SEED)
    nx.draw_networkx_nodes(G, pos, node_size=50)
    nx.draw_networkx_edges(G, pos, alpha=0.35, width=0.8)
    plt.title(f"RAW |{os.path.basename(f)} | N={stats['N']} E={stats['edges']} dens={stats['density']:.3f}")
    plt.axis("off")
    plt.show()
```

```python
# built MCL diffusion-based

def _normalize_columns(M, eps=1e-12):

    col_sums = M.sum(axis=0, keepdims=True)
    col_sums = np.maximum(col_sums, eps)

    return M / col_sums


def _add_self_loops(A, loop_weight=1.0):

    A2 =A.copy()

    np.fill_diagonal(A2, np.maximum(np.diag(A2), loop_weight))
    return A2


def _prune(M, prune_threshold=1e-3):

    M2 = M.copy()

    M2[M2 < prune_threshold] = 0.0
    return M2


def mcl(A, inflation=2.0, expansion=2, loop_weight=1.0,
        prune_threshold=1e-3, max_iter=100, tol=1e-6, verbose=False):
    """
    Markov Clustering:
    - bulding column-stochastic flow matrix
    - repeat expansion + inflation (+ pruning) until convergence

    """
    M = _add_self_loops(A, loop_weight=loop_weight).astype(float)

    M = _normalize_columns(M)

    last = None
    for it in range(1, max_iter + 1):

        #expansion (power)
        Me = M.copy()
        for _ in range(expansion - 1):
            Me = Me @ M

        #inflation
        Mi = np.power(Me, inflation)
        Mi = _normalize_columns(Mi)

        #prune andn renormalize
        Mi = _prune(Mi, prune_threshold=prune_threshold)
        Mi = _normalize_columns(Mi)

        if last is not None:
            diff = np.linalg.norm(Mi - last,ord="fro")

            if verbose:
                print(f"iter={it:03d}, diff={diff:.6e}")
            
            if diff < tol:
                return Mi, it

        last = Mi
        M = Mi

    return M, max_iter


def mcl_clusters_from_matrix(M, eps=1e-6):
    
    """
    extracition of clusters from final MCL matrix:
    - create support graph where M[i,j] > eps (symmetrized)
    - connected components are clusters
    """
    n = M.shape[0]

    S = (M > eps).astype(int)

    S = ((S + S.T) > 0).astype(int)

    G =nx.Graph()

    G.add_nodes_from(range(n))

    r, c = np.where(np.triu(S, 1) > 0)
    for i, j in zip(r, c):
        G.add_edge(i, j)

    comps = list(nx.connected_components(G))

    comps = sorted(comps, key=lambda cc: (len(cc), min(cc)), reverse=True)

    labels = np.zeros(n, dtype=int)
    for cid, comp in enumerate(comps, start=1):
        for v in comp:
            labels[v] = cid

    # just a precaution for isolated nodes
    next_id = labels.max() + 1
    for i in range(n):
        if labels[i] == 0:
            labels[i] = next_id
            next_id += 1

    return labels


def run_mcl_get_labels(A, inflation=2.0, expansion=2, loop_weight=1.0,
                       prune_threshold=1e-3, max_iter=100, tol=1e-6, verbose=False):
    
    M, iters = mcl(
        A, inflation=inflation, expansion=expansion, loop_weight=loop_weight,
        prune_threshold=prune_threshold, max_iter=max_iter, tol=tol, verbose=verbose
    
    )
    
    labels = mcl_clusters_from_matrix(M)
    return labels, iters
```

```python
# forcefully extract K for KNOWN_K by merging clusters so I get always correct K

def _renumber_consecutive(labels):

    uniq = sorted(np.unique(labels))
    remap = {old: new for new, old in enumerate(uniq, start=1)}

    return np.array([remap[x] for x in labels], dtype=int)


def cluster_connectivity_matrix(A, labels):
    """
    C[p,q]=total weight between cluster p and q
    
    """
    labels = np.asarray(labels)
    uniq = sorted(np.unique(labels))
    idx = {c:i for i, c in enumerate(uniq)}
    k = len(uniq)

    C = np.zeros((k, k), dtype=float)


    for i in range(A.shape[0]):
        ci = idx[labels[i]]
        for j in range(A.shape[1]):
            cj = idx[labels[j]]
            C[ci, cj] += A[i, j]

    return C, uniq


def merge_until_k(A, labels, K_target):
    """
    do the merging until the number of clusters equals K_target,
    merge the pair with the largest between-cluster connectivity
    """
    labels =_renumber_consecutive(labels)

    while len(np.unique(labels)) > K_target:
        C, uniq = cluster_connectivity_matrix(A, labels)
        k = len(uniq)

        best_pair = None
        best_val = -1.0

        # choose best non-diagonal pair
        for i in range(k):
            for j in range(i + 1, k):
                val = C[i, j] + C[j, i]
                if val > best_val:
                    best_val = val
                    best_pair = (uniq[i], uniq[j])

        c1, c2 = best_pair
        labels[labels == c2] = c1
        labels = _renumber_consecutive(labels)

    return labels
```

```python
def scan_inflations(A, inflations, expansion=2, loop_weight=1.0,
                    prune_threshold=1e-3, max_iter=100, tol=1e-6):
    
    rows = []
    for r in inflations:
        labels, iters = run_mcl_get_labels(
            A, inflation=r, expansion=expansion, loop_weight=loop_weight,
            prune_threshold=prune_threshold, max_iter=max_iter, tol=tol, verbose=False
        )
        
        k_found = len(np.unique(labels))
        
        rows.append((float(r), int(k_found), int(iters)))
    return pd.DataFrame(rows, columns=["inflation", "K_found", "iters"])


def choose_inflation_known_k(scan_df, K_target):
    df = scan_df.copy()
    
    df["err"] = (df["K_found"] - K_target).abs()
    df = df.sort_values(["err", "inflation"], ascending=[True, True])
    return float(df.iloc[0]["inflation"])


def choose_inflation_unc(scan_df):
    """
    stability basic intuition:
    - find longest consecutive region of constant K_found
    - choose middle inflation from that region
    fallback: is median K_found

    """
    df = scan_df.sort_values("inflation").reset_index(drop=True)
    
    Ks = df["K_found"].values
    
    infls = df["inflation"].values

    best_len = 1
    best_start = 0
    cur_start = 0

    for i in range(1, len(Ks)):
        if Ks[i] != Ks[i - 1]:

            cur_len = i -cur_start
            if cur_len > best_len:
                best_len = cur_len
                best_start = cur_start
            cur_start = i

    cur_len = len(Ks) - cur_start

    if cur_len > best_len:

        best_len = cur_len
        best_start = cur_start

    if best_len >= 2:
        mid = best_start + best_len // 2
        
        return float(infls[mid])

    median_k = int(np.median(Ks))

    candidates= df[df["K_found"] == median_k]
    if len(candidates) > 0:
        return float(candidates.sort_values("inflation").iloc[len(candidates)//2]["inflation"])
    return float(infls[len(infls)//2])


def plot_scan(scan_df, title, K_target=None):
    plt.figure(figsize=(6, 4))
    plt.plot(scan_df["inflation"], scan_df["K_found"], marker="o")
    plt.xlabel("inflation")
    plt.ylabel("K_found")
    plt.title(title)
    plt.grid(True, alpha=0.3)
    if K_target is not None:
        plt.axhline(K_target, linestyle="--")
    plt.show()
```

```python
example_file = files[0]

base = os.path.basename(example_file)
print("Lets try this file: ", base)

A = load_adjacency_matrix(example_file)
cat, K_known = file_category_and_known_k(example_file)

inflations_known = np.round(np.linspace(1.3, 3.2, 11), 2)
inflations_unc   = np.round(np.linspace(1.5, 3.5, 11), 2)


# scan and plot
if cat == "KNOWN_K":
    scan_df = scan_inflations(A, inflations_known)
    plot_scan(scan_df, f"SCAN | {base} | target K={K_known}", K_target=K_known)
    chosen_r = choose_inflation_known_k(scan_df, K_known)
else:
    scan_df = scan_inflations(A, inflations_unc)
    plot_scan(scan_df, f"SCAN | {base} | UNC", K_target=None)
    chosen_r = choose_inflation_unc(scan_df)

print("Chosen inflation:", chosen_r)


#MCL timed
t0 = time.perf_counter()
labels, iters = run_mcl_get_labels(A, inflation=chosen_r)
if cat == "KNOWN_K":
    labels = merge_until_k(A, labels, K_known)  # enforce exact K cause otherwise wrong:(
t1 = time.perf_counter()

print("iters:", iters)
print("K_found:", len(np.unique(labels)))
print("time(s):", t1 - t0)

#plot clustering
G = adjacency_to_graph(A)
pos = nx.spring_layout(G, seed=SEED)

plt.figure(figsize=(7, 6))
nx.draw_networkx_edges(G, pos, alpha=0.30, width=0.8)
nx.draw_networkx_nodes(G, pos, node_size=70, node_color=labels)
plt.title(f"CLUSTERING | {base} | infl={chosen_r} | K={len(np.unique(labels))}")
plt.axis("off")
plt.show()




out_path = os.path.join(OUT_DIR, base)
df_out = pd.DataFrame({0: np.arange(1, A.shape[0] + 1), 1: labels})
df_out.to_csv(out_path, header=False, index=False)

print("Saved:", out_path)
```

```python
#works so lets go with all files

all_results = []  #(file_base, exec_time, chosen_inflation, K_found, category, K_known)

inflations_known = np.round(np.linspace(1.3, 3.2, 11), 2)
inflations_unc   = np.round(np.linspace(1.5, 3.5, 11), 2)


for f in files:
    base = os.path.basename(f)
    A = load_adjacency_matrix(f)
    cat, K_known = file_category_and_known_k(f)

    # scan and plot
    if cat == "KNOWN_K":
        scan_df = scan_inflations(A, inflations_known)
        plot_scan(scan_df, f"SCAN | {base} | target K={K_known}", K_target=K_known)
        chosen_r = choose_inflation_known_k(scan_df, K_known)
    else:
        scan_df = scan_inflations(A, inflations_unc)
        plot_scan(scan_df, f"SCAN | {base} | UNC")
        chosen_r = choose_inflation_unc(scan_df)

    #final run timed
    t0 = time.perf_counter()

    labels, iters = run_mcl_get_labels(A, inflation=chosen_r)

    if cat == "KNOWN_K":

        labels = merge_until_k(A, labels, K_known)
    t1 = time.perf_counter()


    exec_time = t1 - t0
    k_found = int(len(np.unique(labels)))



    out_path = os.path.join(OUT_DIR, base)
    df_out = pd.DataFrame({0: np.arange(1, A.shape[0] + 1), 1: labels})
    df_out.to_csv(out_path, header=False, index=False)


    print(f"Saved {base} | cat={cat} | chosen_r={chosen_r} | K_found={k_found} | time={exec_time:.4f}s")

    # plot final
    G = adjacency_to_graph(A)
    pos = nx.spring_layout(G, seed=SEED)

    plt.figure(figsize=(7, 6))
    nx.draw_networkx_edges(G, pos, alpha=0.30, width=0.8)
    nx.draw_networkx_nodes(G, pos, node_size=70, node_color=labels)
    extra = f"targetK={K_known}" if cat == "KNOWN_K" else "UNC"
    plt.title(f"FINAL | {base} | {extra} | infl={chosen_r} | K={k_found}")
    plt.axis("off")
    plt.show()

    all_results.append((base, exec_time, chosen_r, k_found, cat, K_known))

print("\nAll done:) Outputs saved to:", OUT_DIR)
```
